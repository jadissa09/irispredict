```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris.target_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names) 
```

```python
df[:200]
```

```python
df['target']=iris.target
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
x=df.drop(['target'],axis=1).values
y=df[['target']].values

```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
xtrain=m.fit_transform(xtrain)
xtest=m.transform(xtest)
```

```python
import pickle
pickle.dump(m,open('scaling.pkl','wb'))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

penalty=['l1','l2']
max_iter=[80,100,140]
C=np.linspace(0.1,1,num=5)
```

```python
param_grid=dict(max_iter=max_iter,C=C,penalty=penalty)
liris=LogisticRegression(multi_class="multinomial",solver='saga')
grid=GridSearchCV(estimator=liris,param_grid=param_grid,cv=5)
grid_result=grid.fit(xtrain,ytrain)
grid_result.best_score_,grid_result.best_params_
```

```python
model=LogisticRegression(multi_class="multinomial",solver='saga',C= 0.775, max_iter= 80, penalty= 'l1')
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred)
newdatapred=[[6,3,5,2]]
newdatascaler=m.transform(newdatapred)
model.predict(newdatascaler)
```

```python
import pickle
pickle.dump(model,open('irisreg.pkl','wb')) 
```
